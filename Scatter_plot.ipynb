```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# A simple scatter plot
x=np.linspace(-10,10,50)
x
```

```python
y=np.random.randint(0,300,50)
y
```

```python
z=10*x+3+y
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,z)
```

```python
plt.scatter(y,z)
```

```python
#plot a scatter on pandas data
df=pd.read_csv("batter.csv")
df
```

```python
df2=df.head(50)
df2
```

```python
#scatter plot of pandas data
#avg and strikerate analysis of top 50 player
plt.scatter(df2["avg"],df2["strike_rate"],color="green")
plt.title("average and strike rate  analysis of top 50 batsman")
plt.xlabel("average",color="red")
plt.ylabel("strike rate",color="red")
```

all parameter value is working here

```python
tips=sns.load_dataset("tips")
tips
```

```python
plt.scatter(tips["total_bill"],tips["tip"],marker=">",color="green")
plt.xlabel("total bill")
plt.ylabel("tips")
```

```python
plt.scatter(tips["total_bill"],tips["tip"],s=tips["size"]*10)
# the scatter value is depends upon the size of tips
#s means size
plt.xlabel("total bill")
plt.ylabel("tips")
```

```python
plt.plot(tips["total_bill"],tips["tip"],"o",color="red",marker="+")
```
