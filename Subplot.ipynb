# **Subplot**

multiple graph in a single plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batter=pd.read_csv("batter.csv")
batter.head()
```

```python
# a different way to plot graph
plt.figure(figsize=(18,10))
plt.scatter(batter["avg"],batter["strike_rate"])
plt.xlabel("average")
plt.ylabel("strike_rate")
plt.title("average and strike rate analysis of player")
```

```python
#2nd way of analysis
fig,ax=plt.subplots(figsize=(12,9))
ax.scatter(batter["avg"],batter["strike_rate"],marker="+",color="green")

plt.xlabel("average")
plt.ylabel("strike rate")

fig.show()
```

```python
#creating a multiple graph in a single plot
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(8,6))

ax[0].scatter(batter["runs"],batter["avg"],marker="*",color="blue")
ax[1].scatter(batter["runs"],batter["strike_rate"],marker="+",color="green")

ax[0].set_xlabel("runs")
ax[1].set_xlabel("runs")
ax[0].set_ylabel("average")
ax[1].set_ylabel("strike rate")

ax[0].set_title("runs vs average analysis")
ax[1].set_title("runs vs strike_rate analysis")
```

```python
#using slysing
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,8))

ax[0,0].scatter(batter["runs"],batter["avg"],marker="+",color="red")
ax[0,1].scatter(batter["runs"],batter["strike_rate"],marker="+",color="green")
ax[1,0].hist(batter["strike_rate"],width=15,log=True)      # lograthmic histogram plotted for batter analysis
ax[1,1].hist(batter["avg"],width=4,color="blue")
plt.savefig("sample.png")
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
z=[32,66,43,21,28]
colors=["blue","green","blue","orange","maroon"]
```

```python
# using one by one
fig=plt.figure()

ax1=fig.add_subplot(2,2,1)
ax1.scatter(batter["runs"],batter["avg"])

ax2=fig.add_subplot(2,2,2)
ax2.bar(x,y,color=colors[1::])

ax3=fig.add_subplot(2,2,3)
ax3.pie(z,colors=colors,autopct="%0.1f%%")
plt.show()
```
