```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Colored scatterplot**

```python
iris=pd.read_csv("iris.csv")
iris.sample(5)
```

```python
iris["Species"]=iris["Species"].replace({"Iris-setosa":0,"Iris-virginica":1,"Iris-versicolor":2})
iris.sample(5)
```

```python
#scatter plotting
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Species"],alpha=0.8,cmap="jet")
plt.xlabel("sepel length")
plt.ylabel("petal length")
plt.colorbar()
```

**Annotation**

```python
batter=pd.read_csv("batter.csv")
batter.head()
```

```python
batter_df=batter.head(100).sample(25)
batter_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(batter_df["avg"],batter_df["strike_rate"],s=batter_df["runs"])

for i in range(batter_df.shape[0]):
  plt.text(batter_df["avg"].values[i],batter_df["strike_rate"].values[i],batter_df["batter"].values[i])
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,"point 1")
plt.text(2,6,"point 2")
plt.text(3,7,"point 3")
plt.text(4,8,"point 4",color="red",size="16")
```

**horizontal and vertical lines filtering**

```python
plt.figure(figsize=(18,10))
plt.scatter(batter_df["avg"],batter_df["strike_rate"],s=batter_df["runs"])

plt.axhline(110,color="green",linewidth=7)
plt.axhline(130,color="maroon",linewidth=7)
plt.axvline(30,color="blue",linewidth=7)

plt.xlabel("average of player")
plt.ylabel("strike rate of player")

for i in range(batter_df.shape[0]):
  plt.text(batter_df["avg"].values[i],batter_df["strike_rate"].values[i],batter_df["batter"].values[i])
```
