```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#**3D Plot**

**3D scatter plot**

```python
batter=pd.read_csv("batter.csv")
batter.head()
```

```python
fig = plt.figure()
ax=plt.subplot(projection="3d")

ax.scatter3D(batter["strike_rate"],batter["avg"],batter["runs"],color="green",marker="*")
ax.set_xlabel("strike rate",color="blue")
ax.set_ylabel("average rate",color="blue")
ax.set_zlabel("total runs",color="blue")
ax.set_title("batsman dataset analysis",color="red")
```

**scatter and line both plot**

```python
#merge two graphs in a single plot
fig=plt.figure(figsize=(8,10))
ax=plt.subplot(projection="3d")

x=[0,1,5,2]
y=[6,3,8,7]
z=[8,9,5,7]
ax.scatter3D(x,y,z,s=[100,100,100,100],alpha=0.7,color="orange")
ax.plot3D(x,y,z,color="green")
```

3D surface plot

```python
a = np.linspace(-10,10,100)
b = np.linspace(-10,10,100)
b
```

```python
aa,bb = np.meshgrid(a,b)
aa.shape
```

```python
bb
```

```python
cc = aa**2 + bb**2
cc.shape
```

```python
fig=plt.figure(figsize=(10,8))
ax = plt.subplot(projection ="3d")
p=ax.plot_surface(aa,bb,cc,cmap="viridis")
fig.colorbar(p)
```

```python
c = np.sin(aa)+np.sin(bb)
fig=plt.figure(figsize=(10,8))
ax = plt.subplot(projection ="3d")
p=ax.plot_surface(aa,bb,c,cmap="viridis")
fig.colorbar(p)
```

```python
d = np.log(aa)+np.log(bb)+np.log(aa*bb)
fig=plt.figure(figsize=(10,8))
ax = plt.subplot(projection ="3d")
p=ax.plot_surface(aa,bb,d,cmap="viridis")
fig.colorbar(p)
```

```python
fig=plt.figure(figsize=(8,6))
ax = plt.subplot(projection ="3d")
p=ax.plot_surface(aa,bb,cc,cmap="viridis")
fig.colorbar(p)
```

**contour plot**

**Heatmap**

```python
delevery=pd.read_csv("IPL_Ball_by_Ball_2008_2022.csv")
delevery.head()
```

```python
tempdf=delevery[delevery["ballnumber"].isin([1,2,3,4,5,6]) & (delevery["total_run"]==6)]
tempdf
```

```python
grid=tempdf.pivot_table(index="overs",columns="ballnumber",values="batsman_run",aggfunc="count")
grid
```

```python
fig=plt.figure(figsize=(10,8))
plt.imshow(grid)
plt.yticks(delevery["overs"].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
```
