```python
sk=pd.read_csv("sharma-kohli.csv")
sk
```

```python
plt.plot(sk["RG Sharma"],sk["V Kohli"])
plt.title("rohit and virat ka comparision",color="red")
plt.xlabel("sharma",color="blue")
plt.ylabel("Kohle",color="blue")
plt.show()
```

```python
#ploting multiple graphs
plt.plot(sk["index"],sk["V Kohli"])
plt.plot(sk["index"],sk["RG Sharma"],color="#faaebb")
#hexvalue is available in color parameter
plt.title("virat and rohit carrier comparision")
plt.xlabel("season")
plt.ylabel("runs")
```

```python
#linestyle is a another parameter(dashed,doted,dashdot)
plt.plot(sk["index"],sk["V Kohli"],linestyle="dashed")
plt.plot(sk["index"],sk["RG Sharma"],color="#faaebb",linestyle="dotted")
plt.title("virat and rohit carrier comparision")
plt.xlabel("season")
plt.ylabel("runs")
```

```python
#width of line
plt.plot(sk["index"],sk["V Kohli"],linewidth=4)
plt.plot(sk["index"],sk["RG Sharma"],color="#faaebb",linewidth=5)
plt.title("virat and rohit carrier comparision")
plt.xlabel("season")
plt.ylabel("runs")
```

```python
#marker point(D-diamond,+,dot,>do) and markersize is a mathmatic value
plt.plot(sk["index"],sk["V Kohli"],marker="+",markersize=10)
plt.plot(sk["index"],sk["RG Sharma"],marker="D")
plt.title("virat and rohit carrier comparision")
plt.xlabel("season")
plt.ylabel("runs")
```

```python
plt.plot(sk["index"],sk["RG Sharma"],label="Rohit")
plt.plot(sk["index"],sk["V Kohli"],label="Virat")

plt.legend(loc="best")
#it means legend location it another value is (upper right,lower left, etc)
```
