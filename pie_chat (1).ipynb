```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("fivethirtyeight")
```

# **PIE chat**

.>  univarient or bivarient analysis           
.>  categorical vs numerical column                  
.>  use case --> to find contribution on a standard scale

```python
data=[15,50,48,95,42,120]
subjects=["english","maths","drowing","physics","chemistry","hindi"]
plt.pie(data)
```

```python
plt.pie(data,labels=subjects)
```

```python
colors=["red","green","pink","blue","orange","brown"]
```

```python
plt.pie(data,colors=[x for x in colors],labels=subjects)
plt.show()
```

```python
df=pd.read_csv("gayle-175.csv")

df.head()
```

```python
#fixed the size of graph
plt.figure(figsize=(25,15))
#ploting pie chat using dataset
plt.pie(df["batsman_runs"],labels=df["batsman"],colors=[x for x in colors])
plt.show()
```

```python
df.shape
```

```python
plt.pie(df["batsman_runs"],labels=df["batsman"],colors=[x for x in colors],autopct="%0.5f%%")
plt.show()
```

```python
explval=[0.3,0,0,0,0.2,0]
```

```python
#epload and shadow
plt.pie(df["batsman_runs"],labels=df["batsman"],explode=explval,autopct="%0.5f%%",shadow=True)
plt.show()
plt.style.use("dark_background")
```
