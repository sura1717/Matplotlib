```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#plot simple bar chat
children=[10,20,30,40,50]
color=["yellow","pink","red","blue","green"]
plt.bar(children,color)
```

```python
plt.bar(color,children,color=[x for x in color])
plt.xlabel("colors",color="darkgreen")
plt.ylabel("childrens ",color="darkgreen")
plt.title("colors barchat according to children like",color="darkgreen")
```

```python
#horizeontal bar chat
plt.barh(color,children)
```
